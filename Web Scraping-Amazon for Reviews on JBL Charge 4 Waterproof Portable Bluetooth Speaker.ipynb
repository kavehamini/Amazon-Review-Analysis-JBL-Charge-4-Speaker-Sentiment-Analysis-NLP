## Web Scraping-Amazon for Reviews on JBL Charge 4 Waterproof Portable Bluetooth Speaker 

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing required modules
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
from bs4 import BeautifulSoup
import requests
import re
import pymongo
import pandas as pd
from sqlalchemy import create_engine,inspect, func
import numpy as np
import time
import sqlite3
from datetime import datetime
from pymongo import MongoClient
```

```python
# This is to initialize Splinter for Windows users
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# list to store extracted values

rating_list=[]
date_list=[]
review_title_list=[]
color_list=[]
location_list=[]
comment_list=[]
review_list=[]
locationanddate_list=[]


# loop to scrap website
#scraping 166 pages of reviews
for page_num in range(0, 166, 1):
    url = "https://www.amazon.com/JBL-Waterproof-Portable-Bluetooth-Speaker/product-reviews/B07HKQ6YGX/ref=cm_cr_arp_d_paging_btm_next_"+str(page_num)+"?pageNumber="+str(page_num)
    browser.visit(url)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    results = soup.find_all('div', class_='a-section review aok-relative')


    for result in results:
               
        try:
                
            # scrape the rating
            rating = result.find('a', class_='a-link-normal').i.span.text
            rating_list.append(rating)
            # scrape the  date and location
            locationanddate = result.find('span', class_='a-size-base a-color-secondary review-date').text
            locationanddate_list.append(locationanddate)
                
            date= locationanddate.split(' ')[6:]
            location= locationanddate.split(' ')[3:5]
            
            date_list.append(date)
            location_list.append(location)

            
            
        # scrape the  color
            color = result.find('div', class_='a-row a-spacing-mini review-data review-format-strip').a.text
            color_list.append(color)
            
            
            
            
            
        #scrape the review title
            review_title = result.find('a', class_='a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold').span.text 
            review_title_list.append(review_title)
        #scrape the review
            review = result.find('span', class_='a-size-base review-text review-text-content').span.text
            review_list.append(review)
            
        #scrape the review
            comment= result.find('span', class_='a-expander-prompt').span.text
            comment_list.append(comment)
        except AttributeError:
            date=''
            
            
# store the extracted valuse in dictionary            
Dict={}

Dict["rating"]=rating_list
Dict["date"]=date_list
Dict["location"]=location_list
#Dict["locationanddate"]=locationanddate_list
Dict["color"]=color_list
Dict["review_title"]=review_title_list
Dict["review"]=review_list
Dict["comment"]=comment_list             
```

```python
df=pd.DataFrame.from_dict(Dict, orient='index')
df=df.transpose()
df
```

```python
df['date'] = df['date'].str.join(',')
df['location'] = df['location'].str.join(',')
```

```python
df['date']=df['date'].str.replace(",","-")
df['date']=df['date'].str.replace("--","-")
df['location']=df['location'].str.replace(","," ")
df['rating']=df['rating'].str.replace("out of 5 stars","")
```
